```python
def minibatch_discrimination(activation, batch_size, B, C):
    m = F.reshape(activation, (-1, B, C))
    m = F.expand_dims(m, 3)
    m_T = F.transpose(m, (3, 1, 2, 0))
    m, m_T = F.broadcast(m, m_T)
    l1_norm = F.sum(F.absolute(m-m_T), axis=2)
    
    eraser = F.expand_dims(np.eye(batch_size, dtype="f"), 1)
    eraser = F.broadcast_to(eraser, (batch_size, B, batch_size))
    
    o_X = F.sum(F.exp(-(l1_norm + 1e6 * eraser)), axis=2)
    
    return o_X
```

```python
activation = Variable(np.arange(N*B*C).reshape(-1, B, C).astype("f"))
```

```python
minibatch_discrimination(activation, N, B, C)
```

```python
activation = Variable(np.ones(N*B*C).reshape(-1, B, C).astype("f"))

minibatch_discrimination(activation, N, B, C)
```
